```python
import torch
import math
from torch import Tensor
import matplotlib.pyplot as plt 

# imports from this same folder
import loadData
import baseline
import optim
import modules
import trainAndTest
```

```python
torch.manual_seed(40)
print('Pytorch version: ',torch.__version__)
PLOT_FOLDER = 'plot_folder/'
```

```python
# Generate data with target 0 if outside the circle or 1 if inside
train_input, train_target = loadData.generate_disc_set(1000)
test_input, test_target = loadData.generate_disc_set(1000)

# see if data are close to be balanced and plot the dataset
loadData.plot_data(train_input, train_target, show_plot = False)

# check what accuracy we could expect with a simple baseline
# e.g. with a linear classifier after having added the L2 norm of the points
baseline.baseline_linear_model(train_input,train_target,test_input,test_target)

# convert targets to one-hot encoding
train_target = loadData.convert_to_one_hot(train_target)
test_target  = loadData.convert_to_one_hot(test_target)
print('Targets converted to one-hot')

# Normalize inplace the data
loadData.normalize_data(train_input)
loadData.normalize_data(test_input)

    
```

```python
# create the model

#loss = LossBCE()




optimizers = [optim.SGDOptimizer, optim.SGDmomOptimizer, optim.AdamOptimizer, optim.BFGSOptimizer]
losses = [modules.LossMSE, modules.LossBCE]
```

```python
lr = 0.005
loss_lists = []
train_accs = []
learning_rates=[1,0.1,0.01,0.001,0.0001,0.00001]
for (lr,num) in [(0.002,0),(0.0015,1),(0.002,2),(0.002,3)]:
    Lin1 = modules.Linear(2,25)
    Lin2 = modules.Linear(25,25)
    Lin3 = modules.Linear(25,2)
    act1 = modules.ReLU()
    act2 = modules.ReLU()
    act3 = modules.Tanh()
    act4 = modules.Sigmoid()
    layers = [
        Lin1,
        act1,
        Lin2,
        act2,
        Lin3,
        act3]
    loss = modules.LossMSE()
    model = modules.Sequential(layers, loss)
    optimizer = optimizers[num](model, lr) # 
    epochs = 1000
    mini_batch_size = 50
    loss_list, train_acc = trainAndTest.train(model, optimizer, loss, train_input, train_target,\
                                              epochs, mini_batch_size, verbose = False)
    loss_lists.append(loss_list)
    train_accs.append(train_acc)

    print('optimizer: ',optimizer.name)
    print('learning rate: ',lr)
    print('minimum loss: ',min(loss_list))
    print('maximum accuracy: ',max(train_acc))
    print('On test: ', trainAndTest.test(model,test_input,test_target) )
    trainAndTest.plot_loss_accuracy(loss_list, train_acc)
```

```python
# SGD
lr = 0.001
Lin1 = modules.Linear(2,25)
Lin2 = modules.Linear(25,25)
Lin3 = modules.Linear(25,2)
act1 = modules.ReLU()
act2 = modules.ReLU()
act3 = modules.Tanh()
act4 = modules.Sigmoid()
layers = [
    Lin1,
    act1,
    Lin2,
    act2,
    Lin3,
    act3]
loss = modules.LossMSE()
model = modules.Sequential(layers, loss)
optimizer = optimizers[0](model, lr) # 
epochs = 1000
mini_batch_size = 50
loss_list_sgd, train_acc_sgd = trainAndTest.train(model, optimizer, loss, train_input, train_target,\
                                          epochs, mini_batch_size, verbose = False)


print('optimizer: ',optimizer.name)
print('learning rate: ',lr)
print('minimum loss: ',min(loss_list_sgd))
print('maximum accuracy: ',max(train_acc_sgd))
print('On test: ', trainAndTest.test(model,test_input,test_target) )
trainAndTest.plot_loss_accuracy(loss_list_sgd, train_acc_sgd)
```

```python
# adam
lr = 0.002
Lin1 = modules.Linear(2,25)
Lin2 = modules.Linear(25,25)
Lin3 = modules.Linear(25,2)
act1 = modules.ReLU()
act2 = modules.ReLU()
act3 = modules.Tanh()
act4 = modules.Sigmoid()
layers = [
    Lin1,
    act1,
    Lin2,
    act2,
    Lin3,
    act3]
loss = modules.LossMSE()
model = modules.Sequential(layers, loss)
optimizer = optimizers[2](model, lr) # 
epochs = 1000
mini_batch_size = 50
loss_list_adam, train_acc_adam = trainAndTest.train(model, optimizer, loss, train_input, train_target,\
                                          epochs, mini_batch_size, verbose = False)


print('optimizer: ',optimizer.name)
print('learning rate: ',lr)
print('minimum loss: ',min(loss_list_adam))
print('maximum accuracy: ',max(train_acc_adam))
print('On test: ', trainAndTest.test(model,test_input,test_target) )
trainAndTest.plot_loss_accuracy(loss_list_adam, train_acc_adam)
```

```python
# BFGS
lr = 0.3
Lin1 = modules.Linear(2,25)
Lin2 = modules.Linear(25,25)
Lin3 = modules.Linear(25,2)
act1 = modules.ReLU()
act2 = modules.ReLU()
act3 = modules.Tanh()
act4 = modules.Sigmoid()
layers = [
    Lin1,
    act1,
    Lin2,
    act2,
    Lin3,
    act3]
loss = modules.LossMSE()
model = modules.Sequential(layers, loss)
optimizer = optimizers[3](model, lr) # 
epochs = 1000
mini_batch_size = 50

loss_list_BFGS_3, train_acc_BFGS_3 = trainAndTest.train(model, optimizer, loss, train_input, train_target,\
                                          epochs, mini_batch_size, verbose = False)


print('optimizer: ',optimizer.name)
print('learning rate: ',lr)
print('minimum loss: ',min(loss_list_BFGS_3))
print('maximum accuracy: ',max(train_acc_BFGS_3))
print('On test: ', trainAndTest.test(model,test_input,test_target) )
trainAndTest.plot_loss_accuracy(loss_list_BFGS_3, train_acc_BFGS_3)
```

```python
p0 = plt.figure(0)
loss_lists = [loss_list_sgd, loss_list_mom, loss_list_adam, loss_list_BFGS_3]
train_accs = [train_acc_sgd, train_acc_mom, train_acc_adam, train_acc_BFGS_3]
labels=['SGD','SGDmom','Adam','BFGS']
for i,loss_list in enumerate(loss_lists):
    plt.plot(loss_list, label=labels[i])
plt.grid()
plt.legend()
plt.xlabel('Epochs')
plt.xlim(0,1000)
plt.ylabel('Loss')
plt.title('Loss per optimizer')
plt.show()
p0.savefig(PLOT_FOLDER+'loss per optimizer9')
```

```python
p1 = plt.figure(3)
labels=['SGD','SGDmom','Adam','BFGS']
for i,train_acc in enumerate(train_accs):
    plt.plot([x/train_input.size(0)*100 for x in train_acc], label=labels[i])
plt.grid()
plt.legend()
plt.xlabel('Epochs')
plt.xlim(0,1000)
plt.ylabel('Accuracy in %')
plt.title('Accuracy per optimizer')
plt.show()
p1.savefig(PLOT_FOLDER+'accuracy per optimizer9')
```

```python
import pickle
def read(filename):
    with open (filename, 'rb') as fp:
        itemlist = pickle.load(fp)
    return itemlist
def write(filename,itemlist):
    with open(filename, 'wb') as fp:
        pickle.dump(itemlist, fp)
```

```python
my_loss_list = read('losses_BFGS.p')
my_acc_list = read('acc_BFGS.p')
```

```python
losses_array = np.array(my_loss_list)
acc_array = np.array(my_acc_list)

losses_mean=losses_array.mean(0)
losses_std=losses_array.std(0)

acc_mean = acc_array.mean(0)
acc_std = acc_array.std(0)

losses_std.shape
```

```python
losses_mean_adam = read('losses_mean_adam.p')
losses_std_adam = read('losses_std_adam.p')
acc_mean_adam = read('acc_mean_adam.p')
acc_std_adam = read('acc_std_adam.p')

# SGD
losses_mean_SGD = read('losses_mean_SGD.p')
losses_std_SGD = read('losses_std_SGD.p')
acc_mean_SGD = read('acc_mean_SGD.p')
acc_std_SGD = read('acc_std_SGD.p')

# SGD mom 
losses_mean_SGDmom = read('losses_mean_SGDmom.p')
losses_std_SGDmom = read('losses_std_SGDmom.p')
acc_mean_SGDmom = read('acc_mean_SGDmom.p')
acc_std_SGDmom = read('acc_std_SGDmom.p')

# BFGS
losses_mean_BFGS = read('losses_mean_BFGS.p')
losses_std_BFGS = read('losses_std_BFGS.p')
acc_mean_BFGS = read('acc_mean_BFGS.p')
acc_std_BFGS = read('acc_std_BFGS.p')
```

```python
def plot_mean_and_var(mean, std, style, color,transparency,lenght_array=500, labels=[None,None]):
    plt.plot(range(lenght_array),mean,style+color[0], label=labels[0])
    plt.fill_between(range(lenght_array), mean+(std*1), mean-std, facecolor =color[1], alpha=transparency,label=labels[1])
    return plt
```

```python
fig = plt.figure(figsize=(10,5))
losses_to_plot = [(losses_mean_SGD,losses_std_SGD),(losses_mean_SGDmom,losses_std_SGDmom),\
                 (losses_mean_adam,losses_std_adam),(losses_mean_BFGS,losses_std_BFGS)]
accs_to_plot = [(acc_mean_SGD,acc_std_SGD),(acc_mean_SGDmom,acc_std_SGDmom),\
                 (acc_mean_adam,acc_std_adam),(acc_mean_BFGS,acc_std_BFGS)]

colors = ['r','b','g','c']
opts = ['SGD', 'SGDmom','adam','BFGS']
for i,(mean, std) in enumerate(losses_to_plot):
    plot_mean_and_var(mean[0:500],std[0:500],':',[colors[i],colors[i]],transparency=0.3, labels=['mean_'+opts[i],'std_'+opts[i]])
plt.legend()
plt.grid()
plt.legend()
plt.xlabel('Epochs')
plt.xlim(0,500)
plt.xticks(range(0,500,100))
plt.ylabel('Loss')
plt.title('Loss per optimizer')
plt.show()
fig.savefig(PLOT_FOLDER+'Losses vars_500_bis')
```

```python
fig2 = plt.figure(figsize=(10,5))
for i,(mean, std) in enumerate(accs_to_plot):
    plot_mean_and_var(mean[0:500]/10,std[0:500]/10,':',[colors[i],colors[i]],transparency=0.3, labels=['mean_'+opts[i],'std_'+opts[i]])
plt.legend()
plt.grid()
plt.legend()
plt.xlabel('Epochs')
plt.xlim(0,500)
plt.xticks(range(0,500,100))
plt.ylabel('Accuracy in %')
plt.title('Accuracy per optimizer')
plt.show()
fig2.savefig(PLOT_FOLDER+'accuracies vars_500_bis')
```
